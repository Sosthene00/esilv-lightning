```python
from os import urandom

from bitcointx import select_chain_params
from bitcointx.core import b2x, x
from bitcointx.core.key import BIP32Path
from bitcointx.wallet import CCoinExtKey, CCoinExtPubKey, P2PKHCoinAddress, P2SHCoinAddress, P2WPKHCoinAddress, P2WSHCoinAddress
import bitcointx.core.script as script
```

```python
# first generate some randomness

#seed = urandom(32)
seed = bytes.fromhex('575d1d8a52faca349bd06ab4e41c0ee222a68c78ae0548874f39a944dae26399')

print(f"{b2x(seed)}")
```

```python
# generate a new extended key

#select_chain_params('bitcoin/regtest')
xkey = CCoinExtKey.from_seed(seed)

print(f"{xkey}")
```

```python
# derive a new key pair from path

path = "m/0'/0"

child_key = xkey.derive_path(path)

assert child_key.pub.is_fullyvalid() == True

print(f"New key pair is {child_key.priv}\n{child_key.pub}")
```

```python
from bitcointx.core.script import OP_0, OP_1, OP_CHECKMULTISIG, OP_CHECKSIG

# generate different types of address from the same key

print(f"P2PK: {b2x(script.CScript([bytes(child_key.pub), OP_CHECKSIG]))}")
print(f"P2PKH: {P2PKHCoinAddress.from_pubkey(child_key.pub)}")
print(f"P2WPKH: {P2WPKHCoinAddress.from_pubkey(child_key.pub)}")
multisig_script = script.CScript([OP_0, OP_1, bytes(child_key.pub), OP_1, OP_CHECKMULTISIG])
print(f"Redeem Script: {b2x(multisig_script)}")
print(f"P2SH: {P2SHCoinAddress.from_redeemScript(multisig_script)}")
print(f"P2WSH: {P2WSHCoinAddress.from_redeemScript(multisig_script)}")
```
